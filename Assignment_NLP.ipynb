```python
import numpy as np
import pandas as pd
import re
import nltk
import spacy
import string
```

```python
##Loading the test data
full_data = pd.read_csv('/content/1429_1.csv', nrows=5000)
full_data.head()
```

```python
dt = full_data[["reviews.text"]]
dt["reviews.text"] = dt["reviews.text"].astype(str)
from nltk.tokenize import word_tokenize

### Tokenization

nltk.download('punkt')
tokens = dt['reviews.text'].apply(nltk.word_tokenize)
# Print the tokens
print(tokens)
```

```python
full_data.head()
```

```python
##Lower casing
dt["text_lower"] = dt["reviews.text"].str.lower()
dt.head()
```

```python
##Removal of punctuation
# drop the new column created in last cell
dt.drop(["text_lower"], axis=1, inplace=True)

PUNCT_TO_REMOVE = string.punctuation
def remove_punctuation(reviews_text):
    """custom function to remove the punctuation"""
    return reviews_text.translate(str.maketrans('', '', PUNCT_TO_REMOVE))

dt["text_wo_punct"] = dt["reviews.text"].apply(lambda reviews_text: remove_punctuation(reviews_text))
dt.head()
```

```python
##Removal of stopwords
```

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
", ".join(stopwords.words('english'))
```

```python
STOPWORDS = set(stopwords.words('english'))
def remove_stopwords(reviews_text):
    """custom function to remove the stopwords"""
    return " ".join([word for word in str(reviews_text).split() if word not in STOPWORDS])

dt["text_wo_stop"] = dt["text_wo_punct"].apply(lambda reviews_text: remove_stopwords(reviews_text))
dt.head()
```

```python
## Removal of frequent word
```

```python
from collections import Counter
cnt = Counter()
for text in dt["text_wo_stop"].values:
    for word in text.split():
        cnt[word] += 1

cnt.most_common(10)
```

```python
FREQWORDS = set([w for (w, wc) in cnt.most_common(10)])
def remove_freqwords(text):
    """custom function to remove the frequent words"""
    return " ".join([word for word in str(text).split() if word not in FREQWORDS])

dt["text_wo_stopfreq"] = dt["text_wo_stop"].apply(lambda text: remove_freqwords(text))
dt.head()
```

```python
## Removal of rare words
```

```python
# Drop the two columns which are no more needed
dt.drop(["text_wo_punct", "text_wo_stop"], axis=1, inplace=True)

n_rare_words = 10
RAREWORDS = set([w for (w, wc) in cnt.most_common()[:-n_rare_words-1:-1]])
def remove_rarewords(text):
    """custom function to remove the rare words"""
    return " ".join([word for word in str(text).split() if word not in RAREWORDS])

dt["text_wo_stopfreqrare"] = dt["text_wo_stopfreq"].apply(lambda text: remove_rarewords(text))
dt.head()
```

```python
##Lemmatization
```

```python
import nltk
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
def lemmatize_words(text):
    return " ".join([lemmatizer.lemmatize(word) for word in text.split()])

dt["text_lemmatized"] = dt["reviews.text"].apply(lambda text: lemmatize_words(text))
dt.head()
```

```python
## Stemming
```

```python
from nltk.stem.porter import PorterStemmer

# Drop the two columns
dt.drop(["text_wo_stopfreq", "text_wo_stopfreqrare"], axis=1, inplace=True)

stemmer = PorterStemmer()
def stem_words(text):
    return " ".join([stemmer.stem(word) for word in text.split()])

dt["text_stemmed"] = dt["reviews.text"].apply(lambda text: stem_words(text))
dt.head()
```
