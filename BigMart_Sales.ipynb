```python
train_data=train_data.drop(['Item_Identifier','Outlet_Identifier'], axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
label = ['Item_Fat_Content', 'Outlet_Type', 'Outlet_Location_Type', 'Outlet_Size']
for i in label:
    train_data[i] = LE.fit_transform(train_data[i])
    test_data[i] = LE.fit_transform(test_data[i])
train_data.head()
```

```python
train_data=pd.get_dummies(train_data)
test_data=pd.get_dummies(test_data)
```

```python
train_data.head()
```

```python
from sklearn.model_selection  import cross_val_score as CVS
def cross_val(model, X, y, cv):
    scores = CVS(model, X, y, cv=cv)
    print(f'{model} Scores:')
    for i in scores:
        print(round(i,2))
    print(f'Average {model} score: {round(scores.mean(),4)}')
```

As this is a Regression problem, we are using Regressors
# 1. Linear Regression

```python
from sklearn.model_selection import train_test_split
y = train_data['Item_Outlet_Sales']
X = train_data.drop('Item_Outlet_Sales', axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 20)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
LR = LinearRegression()
LR.fit(X_train, y_train)

# Predict
y_predict = LR.predict(X_test)

# Score Metrics for Regression:
LR_MAE = mean_absolute_error(y_test, y_predict)
LR_MSE = mean_squared_error(y_test, y_predict)
LR_R_2 = r2_score(y_test, y_predict)
print(f" Mean Absolute Error: {LR_MAE}\n")
print(f" Squared Mean Squared Error: {np.sqrt(LR_MSE)}\n")
print(f" R^2 Score: {LR_R_2}\n")

# Cross Validation Score check
cross_val(LR,X,y,5)
```

# Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
DT = DecisionTreeRegressor(max_depth=15, min_samples_leaf=100)
DT.fit(X_train, y_train)
y_predict = DT.predict(X_test)

# Score Metrics (This is where it is not complete)
DT_MAE = mean_absolute_error(y_test, y_predict)
DT_MSE = mean_squared_error(y_test, y_predict)
DT_R_2 = r2_score(y_test, y_predict)
print(f" Mean Absolute Error: {DT_MAE}\n")
print(f" Mean Squared Error: {DT_MSE}\n")
print(f" R^2 Score: {DT_R_2}\n")

# Cross Validation Score check
cross_val(DT,X,y,5)
```

# Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
RF = RandomForestRegressor(n_estimators=100, max_depth=6, min_samples_leaf=100, n_jobs=4, random_state=42)

RF.fit(X_train, y_train)

# Predict
y_predict = RF.predict(X_test)

# Score Matrics
RF_MAE = mean_absolute_error(y_test, y_predict)
RF_MSE = mean_squared_error(y_test, y_predict)
RF_R_2 = r2_score(y_test, y_predict)
print(f" Mean Absolute Error: {RF_MAE}\n")
print(f" Mean Squared Error: {RF_MSE}\n")
print(f" R^2 Score: {RF_R_2}\n")
cross_val(RF,X, y, 5)
```

## **XG Boost**

```python
from xgboost import XGBRegressor
XG= XGBRegressor()
XG.fit(X_train, y_train)
y_predict = XG.predict(X_test)
XG_MAE = mean_absolute_error(y_test, y_predict)
XG_MSE = mean_squared_error(y_test, y_predict)
XG_R_2 = r2_score(y_test, y_predict)
print(f" Mean Absolute Error: {XG_MAE}\n")
print(f" Mean Squared Error: {XG_MSE}\n")
print(f" R^2 Score: {XG_R_2}\n")
cross_val(XG,X, y, 5)
```

# KNN Regressor

```python
from sklearn.neighbors import KNeighborsRegressor
knn= KNeighborsRegressor(n_neighbors=7)
knn.fit(X_train, y_train)
y_predict = knn.predict(X_test)
KNN_MAE = mean_absolute_error(y_test, y_predict)
KNN_MSE = mean_squared_error(y_test, y_predict)
KNN_R_2 = r2_score(y_test, y_predict)
print(f" Mean Absolute Error: {KNN_MAE}\n")
print(f" Mean Squared Error: {KNN_MSE}\n")
print(f" R^2 Score: {KNN_R_2}\n")
cross_val(knn,X, y, 5)
```
